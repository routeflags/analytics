```python
import texthero as hero
import pandas as pd

df = pd.read_csv("stores.csv")
```

```python
df['tfidf'] = (
df['message']
.pipe(hero.clean)
.pipe(hero.tfidf)
)
df['kmeans_labels'] = (
df['tfidf']
.pipe(hero.kmeans, n_clusters=4)
.astype(str)
)
df['pca'] = df['tfidf'].pipe(hero.pca)

hero.scatterplot(df,
                'pca',
                color='kmeans_labels',
                title='K-means reviews',
                hover_data=['message'])
```
