```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
plt.rcParams['font.family'] = 'IPAGothic'
plt.style.use('dark_background')
plt.rcParams["font.size"] = 22
df = pd.read_csv('investigation.csv', index_col='調査ID')
```

```python
df
```

```python
female = df[df['性別'] == '女性']['総利用金額']
male = df[df['性別'] == '男性']['総利用金額']
```

```python
plt.title('総利用金額', fontsize=28)
left = np.array(['女性', '男性'])
height = np.array([female.mean(), male.mean()])
```

```python
plt.bar(left, height)
```

```python
stats.ttest_ind(female, male).pvalue < 0.001
```
