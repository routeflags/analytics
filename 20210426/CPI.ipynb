```python
%load_ext nb_black
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
consumer_price_df = pd.read_csv(
    "zny2015s.csv",
    encoding="cp932",
    index_col=0,
    header=0,
    skiprows=[1, 2, 3, 4, 5],
    parse_dates=True,
)
```

```python
consumer_price_df.tail(5)
```

```python
labels = [
    x
    for x in consumer_price_df.loc["2020-01-01"]
    .sort_values(ascending=False)
    .head(10)
    .tail(5)
    .index
]
labels.insert(0, "総合")
plt.rcParams["figure.figsize"] = (20, 10)
sns.set(style="whitegrid", font="IPAGothic", font_scale=1.5)
ax = sns.lineplot(data=consumer_price_df[labels])
ax.axhline(0, color="red")
```
