```python
from rich import print
import twitter
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
api = twitter.Api(consumer_key=os.environ.get('CONSUMER_KEY'),
                  consumer_secret=os.environ.get('CONSUMER_SECRET'),
                  access_token_key=os.environ.get('ACCESS_TOKEN_KEY'),
                  access_token_secret=os.environ.get('ACCESS_TOKEN_SECRET'))
query = 'q={}%20lang%3Aja%20-filter%3Areplies&src=typed_query&count=100'
```

```python
keywords = ['Kポップ', 'Jロック']
results = [api.GetSearch(
    raw_query=query.format(x + ' -RT'))
          for x in keywords]
favorite_counts = [sum(x.favorite_count 
                       for x in response) 
                  for response in results]
plt.rcParams['font.family'] = 'IPAGothic'
plt.rcParams['figure.figsize'] = 15,6
plt.style.use('dark_background')
plt.rcParams['font.size'] = 19
plt.style.use('dark_background')
plt.title("Twitter ファボ", fontsize=24)
left = np.array(keywords)
height = np.array(favorite_counts)
plt.bar(left, height)
```
