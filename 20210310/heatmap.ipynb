```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('ecom_cosmetics_cart_purchase_labels.csv',
                index_col='event_time',
                parse_dates=True)
df = df[df['purchased'] > 0]
df['hour'] = df.index.hour
df['week'] = df.index.day_name()
pt = df.pivot_table(index='week',
                   columns='hour',
                   values='purchased',
                   aggfunc='sum').fillna(0)
weeks = ["Thursday", "Friday", "Saturday", "Sunday", "Monday", "Tuesday", "Wednesday"]
pt = pt.reindex(weeks)
plt.figure(figsize=[18,12])
sns.set(font_scale=1.8)
sns.heatmap(pt, square=True, linewidths=.5)
```
